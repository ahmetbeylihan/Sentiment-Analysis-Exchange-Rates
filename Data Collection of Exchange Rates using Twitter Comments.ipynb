```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3> Data Collection </h3>

```python
class exchange:
    def __init__(self, currency, currency_description, search_param):
        self.currency = currency
        self.currency_description = currency_description
        self.search_param = search_param
```

```python
class exchange:
    def __init__(self, currency, currency_description, search_param):
        self.currency = currency
        self.currency_description = currency_description
        self.search_param = search_param

exchange_list = []
exchange_list.append(exchange('RUB', 'Russian Rouble ruble', 'rouble rub'))
exchange_list.append(exchange('TRY', 'Turkish Lira', 'turkish lira try'))
exchange_list.append(exchange('EUR', 'Euro', 'euro eur'))
exchange_list.append(exchange('GBP', 'Pound', 'pound gbp'))
```

```python
def collect_tweets(exchange):
    tweets_list = []
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(exchange.search_param + ' since:2021-01-01 until:2021-02-01 lang:en').get_items()):
        tweets_list.append([tweet.user.username, tweet.content, tweet.date, tweet.likeCount,  exchange.currency, tweet])
    tweets_df = pd.DataFrame(tweets_list, columns=['Author', 'Text', 'Time', 'Like Count', 'Currency ISO Code', 'Url'])
    return tweets_df
```

```python
all_twit_df = []
for exchange in exchange_list:
    ex_tweets_df = collect_tweets(exchange)
    all_twit_df.append(ex_tweets_df)
df = pd.concat(all_twit_df)
df
```

```python
df.to_csv('exchange_rates_dataset.csv', encoding='utf-8')
```

<h4>Calculating the change of exchange rate from time to the following day</h4>

By formula <img src="https://raw.githubusercontent.com/ahmetbeylihan/Sentiment-Analysis-Exchange-Rates/911275b7dd98d8c1455d963fee81bb8853bcff5a/percentage%20change%20of%20price.png?token=AOCL62JN4ATKQB7YNISS6GLBW6V44" width=200 height=200 />

```python
def calculate_percentage_change_of_exchange_rate(day, previous_day):
    val = (day / previous_day) - 1 
    if val < 0:
        return -1
    elif val > 0:
        return 1
    else:
        return val
```

```python
for (rowNo, rowData) in df.iterrows():
    try:
        day = (pd.to_datetime(rowData.values[2])).date().strftime('%Y-%m-%d')
        df_day = report_dataset[(report_dataset['Date'] == day)]
        
        previous_day = (pd.to_datetime(rowData.values[2])).date() + timedelta(days=-1)
        df_prev_day = report_dataset[(report_dataset['Date'] == previous_day.strftime('%Y-%m-%d'))]
        
        if rowData.values[4] == 'TRY': 
            val_first = df_day['TRY'].values[0];
            val_prev = df_prev_day['TRY'].values[0];
            df.loc[rowNo, 'target'] = calculate_percentage_change_of_exchange_rate(val_first, val_prev)
        elif rowData.values[4] == 'EUR': 
            val_first = df_day['EUR'].values[0];
            val_prev = df_prev_day['EUR'].values[0];
            df.loc[rowNo, 'target'] = calculate_percentage_change_of_exchange_rate(val_first, val_prev)
        elif rowData.values[4] == 'GBP': 
            val_first = df_day['GBP'].values[0];
            val_prev = df_prev_day['GBP'].values[0];
            df.loc[rowNo, 'target'] = calculate_percentage_change_of_exchange_rate(val_first, val_prev)
        elif rowData.values[4] == 'RUB': 
            val_first = df_day['RUB'].values[0];
            val_prev = df_prev_day['RUB'].values[0];
            df.loc[rowNo, 'target'] = calculate_percentage_change_of_exchange_rate(val_first, val_prev)


    except Exception as e:                                                                          
        print("An error occurred : ", e)   
```
